```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sko.GA import GA
%matplotlib inline
```

```python
X = np.linspace(-2, 2, 50)
A, B, C = 2, 3, -1 # Is the true value
Y = (A * X ** 2 + B * X + C) + np.random.rand(len(X))*0.75 #Quadratic equation
plt.plot(X, Y, 'r-')
plt.grid()
```

```python
def func(X, A, B, C):
    return A * X ** 2 + B * X + C 

def obj_func(p):
     A, B, C = p
     cost = np.square(func(X, A, B, C)- Y).sum()/len(X) # Mean Square Error 
     return cost
  
```

```python
ga = GA(func=obj_func, n_dim = 3, size_pop=100, max_iter=5000,  lb=[-2] * 3, ub= [2] * 3)

best_params, cost = ga.run()
print('best_x:', best_params, '\n', 'best_y:', cost)
```

```python
PredictedY = func(X, *best_params)

fig, ax = plt.subplots()

ax.plot(X, Y, 'r-', label = 'True curve')
ax.plot(X, PredictedY, 'g-', label = 'Predicted curve')
plt.legend(loc='upper left')
plt.grid()


plt.show()
```
